## Introduction

The Ames Assesor's Office house has gotten a call from the Iowa local government saying that it needed to cut costs. In looking at its financial reports, they realized that alot of their costs were spent on collecting data. The Ames City assessor also has dilema as these data collection are used by the tax authority to estimate property prices and its appropriate tax levies. 

In this project, we are trying to create a model that can accurately predict house prices in the city of Ames, located within Story County, Iowa United States to allow users to estimate their house prices. Using data we have obtained from the Ames Assesor's Office house and machine learning models, we will assess the predictive performance of Linear, Lasso and Ridge Regression models in finding accurate prices based on pre-selected features. 

The model with the best predictive performance will be used for our Sale Price prediction and uploaded into [Kaggle](https://www.kaggle.com/c/dsi-us-6-project-2-regression-challenge/leaderboard#score) for a competition between the Ames office and the Tax authorities. They have submitted their score of 35,000 and have urged us to submit ours as soon as possible!

```python
# create a function for subplot histogram
def subplot_histograms(dataframe, list_of_columns, xlabels, ylabels, size):
    nrows = int(np.ceil(len(list_of_columns)/2)) # no. rows to populate 
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=size) # plot subplot, state no. of columns, determine figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, create positions for subplots
    
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column], color='pink',edgecolor='violet') # feel free to add more settings
        ax[i].set_title(column)   # set graph title
        ax[i].set_xlabel(xlabels) # set x label
        ax[i].set_ylabel(ylabels) # set y label
        ax[i].set_ylim(0,2050) # set y axis to 2050 (representing total rows in dataframe)   
    if len(list_of_columns) % 2 == 1:  # Turn off odd number of subplots            
        ax[-1].axis('off')
    sns.set(font_scale=1)  
    plt.tight_layout() 
    
```

```python
# create a function for subplot boxplot
def subplot_box(df, list_x_var, y_var,cols,size):
    nrows = int(np.ceil(len(list_x_var)/cols))
    fig, ax = plt.subplots(nrows=nrows, ncols=cols, figsize=size) 
    ax = ax.ravel()
    for i, column_x in enumerate(list_x_var):                  
        sns.boxplot(data=df, x=list_x_var[i], y=y_var , ax=ax[i])       # Plotting scatterplots
        ax[i].set_xticklabels(ax[i].get_xticklabels(),rotation=90)        # Set x ticks labels to a rotation of 90degrees
        ax[i].set_title(f'{y_var} by {list_x_var[i]}', fontsize = 14)           # Set titles
    if len(list_x_var) % cols == 1:                             
        ax[-1].axis('off')
    plt.tight_layout()
```

Assumption: Linear regression models would work better when target variable is normally distributed  

There are four assumptions associated with a linear regression model:  
1. Linearity: The relationship between X and the mean of Y is linear.  
2. Homoscedasticity: The variance of residual is the same for any value of X.  
3. Independence: Observations are independent of each other.

```python
# create function to get RMSE score for a model
def model_rmse_mean_score(model, X_train, y_train, X_test, y_test, X_kaggle_test, cv_val, kaggle_pred_title ):
    
    model = model
    model.fit(X_train, y_train)
    
    # Finding the performance score for Linear Regression Model fitted with X, y training datasets
    train_score = - cross_val_score(
        model,
        X_train,
        y_train,
        cv=cv_val,
        scoring='neg_root_mean_squared_error'
    ).mean()
    print(f'{model} Train Score: '+str(train_score))
    
    # Finding the performance score for Linear Regression Model fitted with X, y holdout datasets
    holdout_score = - cross_val_score(
        model,
        X_test,
        y_test,
        cv=cv_val,
        scoring='neg_root_mean_squared_error'
    ).mean()
    print(f'{model} Holdout Score: '+str(holdout_score))
    preds[kaggle_pred_title] = model.predict(X_kaggle_test)
    print(f"Kaggle Prediction Array: preds['{kaggle_pred_title}']")
    return kaggle_pred_title
```

Each 1-unit increase in X multiplies the expected value of Y by $e^βˆ\$.

For small values of βˆ, approximately e βˆ ≈ 1+βˆ. We can use this for the following approximation for a quick interpretation of the coefficients: 100 · βˆ is the expected percentage change
in Y for a unit increase in X. For instance for βˆ = .06, e
.06 ≈ 1.06, so a 1-unit change in X
corresponds to (approximately) an expected increase in Y of 6%.
[website](https://kenbenoit.net/assets/courses/ME104/logmodels2.pdf)
