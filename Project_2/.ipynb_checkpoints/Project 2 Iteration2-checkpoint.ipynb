### Introduction

In this project, we are trying to create a model that can accurately predict house prices in the city of Ames, located within Story County, Iowa United States. Using data we have obtained from the Ames Assesor's Office house and Python 3.8 , we will assess the predictive performance of Linear, Lasso and Ridge Regression models in finding accurate prices based on selected features. Afterwhich, we will use the model with the best predictive performance to draw predictions on a test set that will then be uploaded into [Kaggle](https://www.kaggle.com/c/dsi-us-6-project-2-regression-challenge/leaderboard#score) for a competition. 

## Goal
Our goal is to find the best machine learning model that can accurately predict house prices for future homebuyers.  
  
Models Assessed: Linear, Lasso and Ridge   
Scoring: Root Mean Squared Error

```python
def find_diff_classification(colname):
    xtracol_test=[]
    xtracol_train=[]
    for x in test_itr1[colname].unique():
        if not x in train_itr1[colname].unique():
            xtracol_test.append(x)
        else:
            for y in train_itr1[colname].unique():
                if not y in test_itr1[colname].unique():
                    xtracol_train.append(y)
                else:
                    pass
            return print(f'extra columns in test: {xtracol_test} \n extra columns in train: {xtracol_train}')
        
```

### Train Test Split

```python
y = train_itr1['SalePrice']
X = train_itr1[['Lot Area', 'Year Built', 'Total Bsmt SF', 'Gr Liv Area',
       'TotRms AbvGrd', 'Yr Sold', 'MS Zoning_C (all)',
       'MS Zoning_FV', 'MS Zoning_I (all)', 'MS Zoning_RH', 'MS Zoning_RL',
       'MS Zoning_RM', 'Neighborhood_Blmngtn', 'Neighborhood_Blueste',
       'Neighborhood_BrDale', 'Neighborhood_BrkSide', 'Neighborhood_ClearCr',
       'Neighborhood_CollgCr', 'Neighborhood_Crawfor', 'Neighborhood_Edwards',
       'Neighborhood_Gilbert', 'Neighborhood_Greens',
       'Neighborhood_IDOTRR', 'Neighborhood_MeadowV',
       'Neighborhood_Mitchel', 'Neighborhood_NAmes', 'Neighborhood_NPkVill',
       'Neighborhood_NWAmes', 'Neighborhood_NoRidge', 'Neighborhood_NridgHt',
       'Neighborhood_OldTown', 'Neighborhood_SWISU', 'Neighborhood_Sawyer',
       'Neighborhood_SawyerW', 'Neighborhood_Somerst', 'Neighborhood_StoneBr',
       'Neighborhood_Timber', 'Neighborhood_Veenker', 'House Style_1.5Fin',
       'House Style_1.5Unf', 'House Style_1Story', 'House Style_2.5Fin',
       'House Style_2.5Unf', 'House Style_2Story', 'House Style_SFoyer',
       'House Style_SLvl', 'Sale Type_COD', 'Sale Type_CWD', 'Sale Type_Con',
       'Sale Type_ConLD', 'Sale Type_ConLI', 'Sale Type_ConLw',
       'Sale Type_New', 'Sale Type_Oth', 'Sale Type_WD ']]
```

```python
X_train, X_holdout, y_train, y_holdout = train_test_split(
    X,
    y,
    random_state=42,
    test_size=0.2
)
```

### Standardization


```python
ss = StandardScaler()
```

```python
X_train_sc = ss.fit_transform(X_train)
X_holdout_sc = ss.transform(X_holdout)
```

### Testing Linear Regression Model

```python
lr_itr1 = LinearRegression()
lr_itr1.fit(X_train, y_train)
```

```python
y_preds_itr1 = lr_itr1.predict(X_holdout)
y_preds_itr1
```

```python
# Plot the model
plt.figure(figsize=(12,9))
#Generate a scatterplot of predicted values versus actual values. 
plt.scatter(y_preds_itr1, y_holdout, s=5, color='skyblue',alpha=0.5)
#plot a line.
plt.plot([0, np.max(y_holdout)],
        [0,np.max(y_holdout)],
        color = 'black')
#tweak title and axis labels
plt.xlabel('Predicted Values: $\hat{y}$', fontsize = 20)
plt.ylabel('Actual Values: $y$', fontsize = 20)
plt.title('Predicted Values vs Actual Values', fontsize = 24);
```

```python
def model_rmse_mean_score(model, X_train, y_train, X_test, y_test, cv_val):
    
    # Finding the performance score for Linear Regression Model fitted with X, y training datasets
    train_score = - cross_val_score(
        model,
        X_train,
        y_train,
        cv=cv_val,
        scoring='neg_root_mean_squared_error'
    ).mean()
    print(f'{model} Train Score: '+str(train_score))
    
    # Finding the performance score for Linear Regression Model fitted with X, y holdout datasets
    holdout_score = - cross_val_score(
        model,
        X_test,
        y_test,
        cv=cv_val,
        scoring='neg_root_mean_squared_error'
    ).mean()
    print(f'{model} Holdout Score: '+str(holdout_score))
```

```python
model_rmse_mean_score(lr_itr1, 
                 X_train, y_train, 
                 X_holdout, y_holdout, 
                 10)
```

```python
best_ridge_model_itr1.fit(X_train_sc,y_train)
```
