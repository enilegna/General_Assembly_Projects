## Introduction

In this project, we are trying to create a model that can accurately predict house prices in the city of Ames, located within Story County, Iowa United States. Using data we have obtained from the Ames Assesor's Office house and machine learning models, we will assess the predictive performance of Linear, Lasso and Ridge Regression models in finding accurate prices based on pre-selected features. 

The model with the best predictive performance will be used for our Sale Price prediction and uploaded into [Kaggle](https://www.kaggle.com/c/dsi-us-6-project-2-regression-challenge/leaderboard#score) for a competition. 

## Problem Statement

Our goal is to find the best machine learning model that can accurately predict house prices for future homebuyers.  
  
Models Assessed: Linear, Lasso and Ridge   
Scoring: Root Mean Squared Error

#### 2a. Dealing with columns that has a lot of  null entries (apply on both Train & Test)

We will be removing columns that have a lot of null entries with the exception of Fence as it is the only column in the entire dataframe addressing the fence feature.

**To be removed:**
1. Pool QC - Pool Quality (2042 null) - too much null
2. Misc Feature - Miscellaneous features of the house (1986 null) - too much null
3. Fireplace Qu - Fireplace quality (1000 null) - too much null and there are already other fireplace features 
4. Alley - Type of Alley acces to property (1911 null) - too much null

**OHE (one hot encoded) into binary classification:**  
1. Fence (1651 null)  - kept as it is the only column addressing fence
We will be changing fence to a OHE variable where 1 is fenced and 0 is not fenced.

```python
#check for values in fence
train['Fence'].value_counts()
```

```python
# understanding the sales price and fence relationship
sns.boxplot(data=train, x='Fence', y='SalePrice')
plt.title('Fence Type vs SalePrice')
```

From this boxplot we can see that houses with GdPrv fences have better sale price than other kinds of fences. With that in mind we might want to input fence as a feature worth including in our machine learning model. 

```python
# it is likely that null values indicates the absence of fence 
train['has Fence'] = train['Fence'].map(lambda x: 1 if type(x)==str  else 0)
test['has Fence'] = train['Fence'].map(lambda x: 1 if type(x)==str else 0)
```

```python
# plot new one hot encoded feature (has fence)
sns.boxplot(data=train, x='has Fence', y='SalePrice')
plt.title('Fence Type vs SalePrice')
```

Our new data shows that houses that has fences have a slightly lower mean sale price than those without fences. 

This is expected because fences is not a common feature in American homes. In a [quora](https://www.quora.com/Why-do-the-majority-of-American-houses-not-have-front-gates) reply, one user explained that "Where I grew up, having any type of fence or gate in the front was considered rude and not neighborly...Somebody who had a front fence or gate was considered strange and paranoid." 

With this data, we conclude that since fences may not be a strong factor to sale price, we will remove it from our dataset.

```python
# drop all features that are no longer required in 2a.
train.drop(axis=1, columns=['Pool QC','Misc Feature','Fireplace Qu','Alley','Fence','has Fence'], inplace=True)
test.drop(axis=1, columns=['Pool QC','Misc Feature','Fireplace Qu','Alley','Fence', 'has Fence'], inplace=True)
```

```python
# create a function for subplot histogram
def subplot_histograms(dataframe, list_of_columns, xlabels, ylabels, size):
    nrows = int(np.ceil(len(list_of_columns)/2)) # no. rows to populate 
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=size) # plot subplot, state no. of columns, determine figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, create positions for subplots
    
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column], color='pink',edgecolor='violet') # feel free to add more settings
        ax[i].set_title(column)   # set graph title
        ax[i].set_xlabel(xlabels) # set x label
        ax[i].set_ylabel(ylabels) # set y label
        ax[i].set_ylim(0,2050) # set y axis to 2050 (representing total rows in dataframe)   
    if len(list_of_columns) % 2 == 1:  # Turn off odd number of subplots            
        ax[-1].axis('off')
    sns.set(font_scale=1)  
    plt.tight_layout() 
    
```

```python
# create a function for subplot boxplot
def subplot_box(df, list_x_var, y_var,cols,size):
    nrows = int(np.ceil(len(list_x_var)/cols))
    fig, ax = plt.subplots(nrows=nrows, ncols=cols, figsize=size) 
    ax = ax.ravel()
    for i, column_x in enumerate(list_x_var):                  
        sns.boxplot(data=df, x=list_x_var[i], y=y_var , ax=ax[i])       # Plotting scatterplots
        ax[i].set_xticklabels(ax[i].get_xticklabels(),rotation=90)        # Set x ticks labels to a rotation of 90degrees
        ax[i].set_title(f'{y_var} by {list_x_var[i]}', fontsize = 14)           # Set titles
    if len(list_x_var) % cols == 1:                             
        ax[-1].axis('off')
    plt.tight_layout()
```

### Testing Linear Regression Model

```python
preds = pd.Series(dtype=float)
```

```python
def model_rmse_mean_score(model, X_train, y_train, X_test, y_test, X_kaggle_test, cv_val, kaggle_pred_title ):
    
    model = model
    model.fit(X_train, y_train)
    
    # Finding the performance score for Linear Regression Model fitted with X, y training datasets
    train_score = - cross_val_score(
        model,
        X_train,
        y_train,
        cv=cv_val,
        scoring='neg_root_mean_squared_error'
    ).mean()
    print(f'{model} Train Score: '+str(train_score))
    
    # Finding the performance score for Linear Regression Model fitted with X, y holdout datasets
    holdout_score = - cross_val_score(
        model,
        X_test,
        y_test,
        cv=cv_val,
        scoring='neg_root_mean_squared_error'
    ).mean()
    print(f'{model} Holdout Score: '+str(holdout_score))
    preds[kaggle_pred_title] = model.predict(X_kaggle_test)
    print(f"Kaggle Prediction Array: preds['{kaggle_pred_title}']")
    return kaggle_pred_title
```

### Testing RidgeCV Model

```python
def ridge_rmse_mean_score(X_train, y_train, X_test, y_test, X_kaggle_test, cv_val, kaggle_pred_title):
    model = RidgeCV(alphas=np.logspace(0,10,200), store_cv_values=True, scoring ='neg_mean_squared_error') 
    model.fit(X_train, y_train)
    best_model = Ridge(model.alpha_)
    best_model.fit(X_train, y_train)
    
    # Finding the performance score for Linear Regression Model fitted with X, y training datasets
    train_score = - cross_val_score(
        best_model,
        X_train,
        y_train,
        cv=cv_val,
        scoring='neg_root_mean_squared_error'
    ).mean()
    print(f'{best_model} Train Score: '+str(train_score))
    
    # Finding the performance score for Linear Regression Model fitted with X, y holdout datasets
    holdout_score = - cross_val_score(
        best_model,
        X_test,
        y_test,
        cv=cv_val,
        scoring='neg_root_mean_squared_error'
    ).mean()
    print(f'{best_model} Holdout Score: '+str(holdout_score))
    preds[kaggle_pred_title] = model.predict(X_kaggle_test)
    print(f"Kaggle Prediction Array: preds['{kaggle_pred_title}']")
```

### Testing Lasso Model

```python
def lasso_rmse_mean_score(X_train, y_train, X_test, y_test, X_kaggle_test, cv_val, kaggle_pred_title):
    model = LassoCV(n_alphas = 500,  cv=10) 
    model.fit(X_train, y_train)
    best_model = Lasso(model.alpha_)
    best_model.fit(X_train, y_train)
    
    # Finding the performance score for Linear Regression Model fitted with X, y training datasets
    train_score = - cross_val_score(
        best_model,
        X_train,
        y_train,
        cv=cv_val,
        scoring='neg_root_mean_squared_error').mean()
    print(f'{best_model} Train Score: '+str(train_score))
    
    # Finding the performance score for Linear Regression Model fitted with X, y holdout datasets
    holdout_score = - cross_val_score(
        best_model,
        X_test,
        y_test,
        cv=cv_val,
        scoring='neg_root_mean_squared_error').mean()

    print(f'{best_model} Holdout Score: '+str(holdout_score))
    preds[kaggle_pred_title] = model.predict(X_kaggle_test)
    print(f"Kaggle Prediction Array: preds['{kaggle_pred_title}']")
    return kaggle_pred_title
```
